```python
def visualize_dataset(dataset):
  from sklearn.preprocessing import StandardScaler
  print(dataset)

  length = len(dataset[0])+len(dataset[1])

  data_tensor = torch.zeros([length, dataset[0].shape[1]], dtype=torch.float32)
  print(data_tensor.shape)

  for i in range(length):
    if i < length/2:
      data_tensor[i] = dataset[0][i]
    else:
      data_tensor[i] = dataset[1][i-length]

  data_normalized = StandardScaler().fit_transform(np.array(data_tensor))

  from sklearn.manifold import TSNE

  tsne = TSNE(n_components=2, perplexity=5.0) # reduce data to 2 components, perplexity must be less than amount of samples (will run tmr)
  print(data_normalized.shape)
  tsne_results = tsne.fit_transform(data_normalized) # applies the t-SNE transformation to the normalized data - 2D representation of the data

  plt.figure(figsize=(10, 8))
  categories = ["Sycophantic activations", "Non sycophantic activations"]
  colors = ['b','r']  # Choose colors that match your desired output

  idx = 0
  for cat, color in zip(categories, colors):
    if idx < len(dataset[0]):
      plt.scatter(data_normalized[idx, 0], data_normalized[idx, 1], c=color, label=cat, marker=".")
    elif idx >= len(dataset[0]) and idx < len(dataset[1]):
      plt.scatter(data_normalized[idx, 0], data_normalized[idx, 1], c=color, label=cat, marker=".")

  plt.legend()
  plt.xlabel('t-SNE 1')
  plt.ylabel('t-SNE 2')
  plt.title('t-SNE Visualization')
  plt.show()
```
