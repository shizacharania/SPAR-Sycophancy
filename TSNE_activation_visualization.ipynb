```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 8))

categories = ["Sycophantic activations - A", "Sycophantic activations - B",
              "Non sycophantic activations - A", "Non sycophantic activations - B"]
colors = ['b', 'b', 'r', 'r']  # Choose colors that match your desired output

# print(labels[0] == categories[0])
# print(indices_list)

indices_list = []
for category in categories:
    current_indices = []
    for i in range(len(labels)):
      if labels[i] == category:
        current_indices.append(i)
    indices_list.append(current_indices)
print(indices_list)

idx = 0
for cat, color in zip(categories, colors):
    if idx % 2 == 0:
      plt.scatter(data_normalized[indices_list[idx], 0], data_normalized[indices_list[idx], 1], c=color, label=cat, marker=".")
    else:
      plt.scatter(data_normalized[indices_list[idx], 0], data_normalized[indices_list[idx], 1], c=color, label=cat, marker="x")
    idx += 1

plt.legend()
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.title('t-SNE Visualization')
plt.show()
```
