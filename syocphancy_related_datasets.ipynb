```python
import random

print(len(empatheticdialogues_train))
confident_score = 0
anxious_score = 0
apprehensive_score = 0

confident_prompts = []
confident_utterances = []
not_confident_prompts = []

for i in range(1, len(empatheticdialogues_train)):
  if empatheticdialogues_train['context'][i] == "confident":
    confident_prompts.append(empatheticdialogues_train['prompt'][i])
    confident_utterances.append(empatheticdialogues_train['utterance'][i])
    confident_score += 1
  elif empatheticdialogues_train['context'][i] == "anxious":
    not_confident_prompts.append(empatheticdialogues_train['prompt'][i])
    anxious_score += 1
  elif empatheticdialogues_train['context'][i] == "apprehensive":
    not_confident_prompts.append(empatheticdialogues_train['prompt'][i])
    apprehensive_score += 1
```
